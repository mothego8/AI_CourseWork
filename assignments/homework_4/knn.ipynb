"""Now let's classify the point as either M or F.  We'll do this by setting k = 1 and using the Euclidean distance.  We'll define that as:"""


```python
def euclideanDistance(instance1, instance2, length):
   distance = 0
   for x in range(length):
         #print ('x is ' , x)
         num1=float(instance1[x])
         num2=float(instance2[x])
         distance = distance + pow(num1-num2, 2)
   return math.sqrt(distance)
```

Manhattan Distance

```python
def manhattanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
         #print ('x is ' , x)
         num1=float(instance1[x])
         num2=float(instance2[x])
         distance += abs(num1-num2)
    return distance
```

Hamming Distance

```python
def hammingDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
         #print ('x is ' , x)
         num1=float(instance1[x])
         num2=float(instance2[x])
         distance += int(num1 != num2) 
    return distance
```

Finds the closest k neighbors and takes mode

```python
# multiple k
def knn(distances, k):
    # Assuming distances is a list of objects with a 'tag' attribute
    labels = [distance.tag for distance in distances[:k]]

    # Find the most common label
    label = max(set(labels), key=labels.count)
    return label
```

SOLUTION

```python
#Add code here

# instantiating variables for all accuracies
euc_accuracy_1 = 0
man_accuracy_1 = 0
ham_accuracy_1 = 0

euc_accuracy_7 = 0
man_accuracy_7 = 0
ham_accuracy_7 = 0

euc_accuracy_15 = 0
man_accuracy_15 = 0
ham_accuracy_15 = 0

# loop through test data
for i in test_data:
    eucDistances2 = [] # list of distances, will hold objects of type distClass
    manDistances2 = []
    hamDistances2 = []

    point = i
    point_label = point[-1]

    # for each point in test data we're going to find its distance to every point in training data
    # and add those distances to their respective lists
    for j in train_data: 
        temp = j
        label = temp[-1]
        length = len(temp) - 1

        d = euclideanDistance(point,temp, length)
        m = manhattanDistance(point, temp, length)
        h = hammingDistance(point, temp, length)

        eucObj = distClass() #one record's distance and tag
        eucObj.dist=d
        eucObj.tag=label
        eucDistances2.append(eucObj)

        manObj = distClass()
        manObj.dist = m
        manObj.tag = label
        manDistances2.append(manObj)

        hamObj = distClass()
        hamObj.dist = h
        hamObj.tag = label
        hamDistances2.append(hamObj)

    # sort the different lists that contain that distances from the test point to each train point
    eucDistances2.sort(key=lambda x: x.dist)
    manDistances2.sort(key=lambda x: x.dist)
    hamDistances2.sort(key=lambda x: x.dist)

    # check if result of knn is same as what tag is supposed to be 
    # k = 1
    if knn(eucDistances2, 1) == point_label:
        euc_accuracy_1 += 1
    if knn(manDistances2, 1) == point_label:
        man_accuracy_1 += 1
    if knn(hamDistances2, 1) == point_label:
        ham_accuracy_1 += 1
    # k = 7
    if knn(eucDistances2, 7) == point_label:
        euc_accuracy_7 += 1
    if knn(manDistances2, 7) == point_label:
        man_accuracy_7 += 1
    if knn(hamDistances2, 7) == point_label:
        ham_accuracy_7 += 1
    # k = 15
    if knn(eucDistances2, 15) == point_label:
        euc_accuracy_15 += 1
    if knn(manDistances2, 15) == point_label:
        man_accuracy_15 += 1
    if knn(hamDistances2, 15) == point_label:
        ham_accuracy_15 += 1


# extract the size of test data
size = len(test_data)

# divide all accuracies by size of test data to find their true accuracy

euc_accuracy_1 /= size
man_accuracy_1 /= size
ham_accuracy_1 /= size

euc_accuracy_7 /= size
man_accuracy_7 /= size
ham_accuracy_7 /= size

euc_accuracy_15 /= size
man_accuracy_15 /= size
ham_accuracy_15 /= size
    
```
